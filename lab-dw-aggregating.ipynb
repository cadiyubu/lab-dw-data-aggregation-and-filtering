In this challenge, we will continue to work with customer data from an insurance company. We will use the dataset called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
import pandas as pd
url='https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'

df1=pd.read_csv(url)
df=df1.copy()
display(df.head(2))
df.info()
```

1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
#DataFrame based on total_claim_amount<1000 and response Yes to marketing campaign
print("null values for total claim amount?")
display(df['Total Claim Amount'].isnull().sum())
print("null values for Response?")
display(df['Response'].isnull().sum())
#Now we can create the DataFrame based on total_claim_amount<1000 and response Yes 
claim_df=df[(df['Response']=='Yes') & (df['Total Claim Amount']<1000) ]
display(claim_df.shape)
claim_df.head(5)

```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
#to see the average monthly_premium and/or customer lifetime value by policy_type and gender for 
#customers who responded "Yes", first we need to check how those columns are formated and their values.
```

```python
display(df['Policy Type'].unique())
display(df['Gender'].unique())
```

```python
#Now we can create a pivot table to see the average monthly_premium and/or customer lifetime value by policy_type and gender for 
#customers who responded "Yes"
col=['Monthly Premium Auto','Customer Lifetime Value']
yes_piv= df[df.Response=='Yes'].pivot_table(index=['Policy Type','Gender'], values=col,aggfunc='mean')
display(yes_piv.round(2))
```

```python
#compare these insights to total_claim_amount patterns
```

```python
#comparing to the DataFrame based on total_claim_amount<1000 and response Yes to marketing campaign
claim_df.head(2)
claim_piv= claim_df.pivot_table(index=['Policy Type','Gender'], values=col,aggfunc='mean')
display(claim_piv.round(2))
```

```python
#comparing to the DataFrame based on total_claim_amount<1000 and response Yes to marketing campaign
pd.concat([yes_piv.round(2),claim_piv.round(2)], axis=1)
```

Interpreting results:

Looking at the groups that responded Yes to the campaign, without a claim amount threshold, the females with the policy Personal Auto are the ones who brind the higher CLV (8339.79 USD). When considering the threshold of claim amount (<1000 USD), it shows that the highest CLV comes from the males with policy type Special Auto (8348.23).

If we look at the CLV with and without the claim amount threshold (<1000 USD), then we can say that the highest CLV comes from the Males who had low claim amount and with Special Auto Policy Type who bring 8348.23 USD of CLV in average, followed by the Females with policy type Personal Auto who bring 8339.79 USD CLV in average. These would be then the most profitable segments.



3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
#Analyze the total number of customers who have policies in each state
#as there are nan values for State, we will not consider those
cust_piv=df[df.State.isnull()==False].pivot_table(index='State',values='Customer',aggfunc='count',fill_value=0,margins=True, margins_name='Total')
cust_piv
```

```python
#from the table above we can see that all states have above 500 customer, being California with the highest
#number of customer, with 3552.
```

```python
#the number of customer who have more than 500 customers are
cust_piv[cust_piv.Customer>500]
```

```python
#Another way to do it as AI reviewer suggests
state_counts=df.groupby('State')['Customer'].count()
display(state_counts)
result=state_counts[state_counts >500]
result
```

```python
#from the table above we can see that all states have above 500 customer, being California with the highest
#number of customer, with 3552.
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
#Getting the customer lifetime value by education level and gender and agg by(maximum, minimum, and median )
df.groupby(['Education', 'Gender'])['Customer Lifetime Value'].agg(['max', 'min', 'median']).round(2)
```

```python
# Observations:
#- The maximum CLV comes from the Men with High School or Below education with CLV = 83325.38 USD and while the maximum from women comes from the ones with the ones with Bachelor with a CLV of 73225.96 USD.
#- The minimum CLV comes from males with Bachelor (1898.01 USD) which is comparable to the women with collegue (1898.68), just cents difference.
#- The highest median comes from the males with High School or Below, with CLV 6286.73 USD and the lowes median comes from Females wit Doctor degree, with CLV of 5332.46 USD.

# Interpretation: 
#CLV does not scale monotonically with education. Education level alone doesn't predict customer value — lower-education groups actually contain the highest-spending outliers, and gender effects depend on which education group you're looking at.
```
